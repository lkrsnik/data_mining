```python
def log(x, e=1e-10):
    return np.log(x+e)

# funkcija za pretvorbo rzredov tipa string v float
def razredeVFloat(primer):
    options = {"Class_1" : 1.0, "Class_2" : 2.0, "Class_3" : 3.0, "Class_4" : 4.0, "Class_5" : 5.0, 
               "Class_6" : 6.0, "Class_7" : 7.0, "Class_8" : 8.0, "Class_9" : 9.0}
    return options[primer]

# funkcija, ki prebere datoteke in vrne neodvisne in odvisne spremenljivke
def preberiVrniUcnePodatke(datoteka):
    indeksiAtributov = [x for x in range(0, 94)]
    atributi = np.genfromtxt(datoteka, skiprows = 1, dtype = float, delimiter=',', usecols = indeksiAtributov)    
    razredi = np.genfromtxt(datoteka, skiprows = 1, dtype = str, delimiter=',', usecols = {94})
    
    funkcijaSpremeni = np.vectorize(razredeVFloat)
    razredi = funkcijaSpremeni(razredi)
    
    return Orange.data.Table(atributi, razredi)

# Sigmoidna funkcija
def sigmoid(X):
    '''Compute the sigmoid function '''
    #d = zeros(shape=(X.shape))
 
    den = 1.0 + np.e ** (-1.0 * X)
 
    d = 1.0 / den
 
    return d

def h(theta, x):
    return 1. / (1 + np.exp(-x.dot(theta)))

# Naslednjo funkcijo bomo minimizirali z uporabo L-BFGS-B algoritma
# http://www.holehouse.org/mlclass/07_Regularization.html
def costFunctionWithRegularization(theta, *args):
    y = args[0] # odvisna spremenljivka
    X = args[1] # neodvisne spremenljivke
    r = args[2] # regularizacijski faktor
    m = len(y)  # stevilo vseh primerov
    thetaR = theta[1:]
    h = sigmoid(X.dot(theta))
    
    # Naslednji dve rešitvi vrneta popolnoma isti rezultat
    #J = (1./m) * (-np.transpose(y).dot(np.log(sigmoid(X.dot(theta)))) - np.transpose(1-y).dot(np.log(1-sigmoid(X.dot(theta))))) + (r / (2 * m)) * (thetaR.T.dot(thetaR)) 
    J = (1.0 / m) * ((-y.T.dot(log(h))) - ((1 - y.T).dot(log(1.0 - h)))) + (r / (2.0 * m)) * (thetaR.T.dot(thetaR))
    return J
    # return (-1/m)*((y.dot(log(h(theta, X))))+(1-y).dot(log(1-h(theta, X))))+(lamb/(2*m))*theta.dot(theta)
    #return (1 / (2 * m)) * (((X.dot(theta) - y).T.dot(X.dot(theta) - y)) + (r * theta.T.dot(theta)))
 
"""
def costFunctionWithRegularization2(theta, *args):
    y = args[0] # odvisna spremenljivka
    X = args[1] # neodvisne spremenljivke
    r = args[2] # regularizacijski faktor
    m = len(y)  # stevilo vseh primerov
    thetaR = theta[1:]
    #h = sigmoid(X.dot(theta))

    J = (-1/m)*((y.dot(log(h(theta, X))))+(1-y).dot(log(1-h(theta, X))))+(r/(2*m))*theta.dot(theta)
    return J
"""

def costFunctionWithRegularizationGradient(theta, *args):
    '''Compute the cost and partial derivatives as grads
    '''
    y = args[0] # odvisna spremenljivka
    X = args[1] # neodvisne spremenljivke
    r = args[2] # regularizacijski faktor
    m = len(y)  # stevilo vseh primerov
    thetaR = theta[1:] # Za vse thete razen 0. - za regularizacijo
    h = sigmoid(X.dot(theta))
    
    # Za 0. theto oz. X
    sumdelta = (h - y).T.dot(X[:, 0])
    grad0 = (1.0 / m) * sumdelta
    
    # Za ostale thete oz. X
    XR = X[:, 1:X.shape[1]]
    sumdelta = (h - y).T.dot(XR)
    grad = (1.0 / m) * (sumdelta + thetaR * r)
    # Gradient
    return np.hstack((grad0, grad)).T.flatten()

"""
def costFunctionWithRegularizationGradient2(theta, *args):
    '''Compute the cost and partial derivatives as grads
    '''
    y = args[0] # odvisna spremenljivka
    X = args[1] # neodvisne spremenljivke
    r = args[2] # regularizacijski faktor
    m = len(y)  # stevilo vseh primerov
    thetaR = theta[1:] # Za vse thete razen 0. - za regularizacijo
    return (1/m)*(((h(theta, X) - y).dot(X))) +((r/m)*theta)
"""

# Funkcija za numericno preverjanje
def numerical_grad(f, params, epsilon):
    #num_grad = np.zeros_like(theteReg)
    num_grad = np.zeros(len(params))
    #perturb = np.zeros_like(params)
    perturb = np.zeros(len(params))
    for i in range(params.size):
        perturb[i] = epsilon
        j1 = f(params + perturb)
        j2 = f(params - perturb)
        num_grad[i] = (j1 - j2) / (2. * epsilon)
        perturb[i] = 0
    return num_grad
```

```python
# pridobim neodvisne in odvisna spremenljivke v Orange.data.Table
ucniPodatki = preberiVrniUcnePodatke("train.csv")
# Spredaj v X domam stolpec 1

print
# In[13]:

vseThete = []
for i in range(1, len(ucniPodatki.domain.class_var.values)):
#for i in range(0, len(ucniPodatki.domain.class_var.values)):
    # priprava podatkov, najprej primeri z razredom i
    X1 = ucniPodatki.X[ucniPodatki.Y == i]
    y1 = np.array([0] * X1.shape[0]) # cilji razred ima indeks 0
    
    
    
    # sledi vzorec primerov z razredom, ki je drugačen od i
    ind = np.arange(len(ucniPodatki))[ucniPodatki.Y != i]
    np.random.shuffle(ind)
    X0 = ucniPodatki.X[ind[:X1.shape[0]]]
    y0 = np.array([1] * X0.shape[0])

    # oblikujemo dvovrednostno domeno
    domain = Orange.data.Domain(ucniPodatki.domain.attributes, 
                                Orange.data.DiscreteVariable("class", values=["T", "F"]))
    # podatke združimo v tabelo
    X = np.vstack((X1, X0))
    # Prvi stolpec nastavim na same 1
    X[:,0] = 1
    #X = np.column_stack((np.ones(len(X)), X))
    y = np.hstack((y1, y0))
    #xy = Orange.data.Table(domain, X, y)
    nicle = np.zeros(X.shape[1]).T
    theteReg = scipy.optimize.fmin_l_bfgs_b(costFunctionWithRegularization, x0=nicle, fprime=costFunctionWithRegularizationGradient,args=(y, X, 0.01))[0]
    vseThete.append(theteReg)
    break
    # tabelo podamo binarnemu učnemu algoritmu in shranimo vrnjeni klasifikator v seznam
    #classifiers.append(self.learner(xy))
# seznam klasifikatorjev podamo objektu, ki bo primere klasificiral
#mc = MulticlassModel(data.domain, classifiers)
#mc.name = self.name
#return mc


# In[14]:

indeksiAtributov = [x for x in range(0, 94)]
testniPrimeri = np.genfromtxt("test.csv", skiprows = 1, dtype = float, delimiter=',', usecols = indeksiAtributov)
```
