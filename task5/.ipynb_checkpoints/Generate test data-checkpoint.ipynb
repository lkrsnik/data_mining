```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def g(z):
    return 1/(1+np.exp(-z))

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
def add_ones(X):
    return np.column_stack((np.ones(len(X)), X))
X = add_ones(X)

# AND
theta = np.array([-30, 20, 20])
y = g(X.dot(theta))

# OR
theta = np.array([-10, 20, 20])

# NOR
theta = np.array([10, -20, -20])

# NOT
theta = np.array([10, -20, 0])

# XOR
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
a1 = add_ones(X)
theta1 = np.array([[-30, 10], [20, -20], [20, -20]])
z2 = a1.dot(theta1)
a2 = add_ones(g(z2))
theta2 = np.array([[-10], [20], [20]])
z3 = a2.dot(theta2)
a3 = g(z3)
y = a3

# FEED FORWARD
class FeedForward:
    def __init__(self, theta1, theta2):
        self.theta1 = theta1
        self.theta2 = theta2
        
    def __call__(self, x):
        a1 = add_ones(x)
        z2 = a1.dot(theta1)
        a2 = add_ones(g(z2))
        z3 = a2.dot(theta2)
        return g(z3)
    
theta1 = np.array([[-25, 15], [20, -20], [20, -20]])
theta2 = np.array([[-10], [20], [20]])
ff = FeedForward(theta1, theta2)

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
ff(X)

n = 50
X = np.array(list(zip(np.random.rand(n), np.random.rand(n))))

cls = [int(yi>=0.5) for yi in ff(X).reshape(-1)]
plt.scatter(X[:, 0], X[:, 1], color=np.array(["g", "b"])[cls]);
```
